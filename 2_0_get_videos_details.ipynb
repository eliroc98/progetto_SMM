```python
import pandas as pd

import requests 

import os

import json

import googleapiclient.discovery
import googleapiclient.errors

import numpy as np
```

```python
df_videos = pd.read_csv('data/get_youtuber_videos_output/youtuber_videos.csv')
```

```python
chiave_yt = json.load(open('key.json', 'r'))['youtube']['key'][2]
youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=chiave_yt)
```

```python
#rimetti a 0 se parti daccapo
quota = 0
```

```python
def get_infos(response):
    title= description= category_id=relevant_topic_ids=relevant_topic_categories = None
    all_details = True
    try:
        snippet = response['items'][0]['snippet']
        title = snippet['title']
        description = snippet['description']
        category_id = snippet['categoryId']
    except KeyError as e:
        all_details = False
        print(e.args[0])
    try:
        topicDetails = response['items'][0]['topicDetails']
        relevant_topic_ids = topicDetails['relevantTopicIds']
        relevant_topic_categories = topicDetails['topicCategories']
    except KeyError as e:
        all_details = False
        print(e.args[0])
    return title, description, category_id,relevant_topic_ids,relevant_topic_categories , all_details
```

```python
last_csv = int(np.load('data/get_videos_details/last_csv.npy'))
last_video_idx = int(np.load('data/get_videos_details/last_video_idx.npy'))
videos_not_complete = (np.load('data/get_videos_details/videos_not_complete.npy')).tolist()
error_videos = (np.load('data/get_videos_details/error_videos.npy')).tolist()
print("Starting with video #{}".format(last_video_idx))
print("Writing on {} csv".format(last_csv+1))
```

```python
#first_round
df_videos_0 = df_videos[last_video_idx:]
df_videos_0
```

```python
videos_details = []
videos_count = 0
```

```python
for  video in df_videos_0['video_id'].values:
    if quota < 9900:
        print('_________________________________________')
        print('{} Video: {}'.format(last_video_idx, video))
        print('Reached quota {}'.format(quota))
        request = youtube.videos().list(
            part="snippet,topicDetails",
            id=video
        )
        try:
            response = request.execute()
            r = response['items'][0]
            quota += 5
            title, description, category_id,relevant_topic_ids,relevant_topic_categories, all_details = get_infos(response)
            if not all_details:
                videos_not_complete.append(video)
            videos_details.append([video,title, description, category_id,relevant_topic_ids,relevant_topic_categories])
            last_video_idx += 1
        except:
            error_videos.append(video)
            print('Something went wrong with video {}'.format(video))
    else:
        print('Quota limit exceeded')
        break
```
