```python
import pandas as pd

import os

import numpy as np

import networkx as nx

import matplotlib.pyplot as plt

from IPython.display import clear_output
```

```python
#importo il grafo completo
first_graph = nx.read_gpickle('data/graphs/positive/positive_graph.gpickle')
```

```python
#rimuovo i singoletti
first_graph.remove_nodes_from(list(nx.isolates(first_graph)))
```

```python
topics = list(dict((np.load('data/topics/topic_ids.npy',allow_pickle=True)).tolist()).values())
[topics.append(item[1]) for item in np.load('data/topics/keywords.npy', allow_pickle=True).tolist()]
[topics.append(item) for item in list(dict((np.load('data/topics/categories_ids.npy',allow_pickle=True)).tolist()).values())]
topic_set = set(topics)
topic_set
```

```python
brands = [item[-1] for item in np.load('data/topics/makeup_brands.npy', allow_pickle=True).tolist() if item[-1]!=""]
brand_set = set(brands)
brand_set
```

```python
df_youtubers = pd.read_csv('data/get_youtubers_output/youtubers.csv')
youtubers_set = set(df_youtubers['display_name'].values)
youtubers_set
```

```python
#preparo il counter
dict_attrs = dict()
for idx, attrs in enumerate(first_graph.nodes.values()):
    for attr in attrs.keys():
        dict_attrs.update({attr:0})
```

```python
for idx, attrs in enumerate(first_graph.nodes.values()):
    print('{}/{}'.format(idx,len(first_graph.nodes)))
    clear_output(wait=True)
    for key_attr, key_value in zip (attrs.keys(),attrs.values()):
        dict_attrs[key_attr] += key_value
```

```python
rank_topics=[]
rank_brands=[]
rank_youtubers=[]
for key_item,key_value in zip (dict_attrs.keys(),dict_attrs.values()):
    if key_item in topic_set:
        rank_topics.append((key_item,key_value))
    if key_item in brand_set:
        rank_brands.append((key_item,key_value))
    if key_item in youtubers_set:
        rank_youtubers.append((key_item,key_value))
```

```python
rank_topics = sorted(rank_topics, key=lambda x:x[1],reverse = True)
rank_youtubers = sorted(rank_youtubers, key=lambda x:x[1],reverse = True)
rank_brands = sorted(rank_brands, key=lambda x:x[1],reverse = True)
```

```python
print("Top 3 topics: {}".format(rank_topics[:3]))
print("Top 3 youtubers: {}".format(rank_youtubers[:3]))
print("Top 3 brands: {}".format(rank_brands[:3]))
```

```python
assortativity_coefficients = np.load('analysis/positive/assortativity_coefficients.npy',allow_pickle=True)
assortativity_coefficients=dict(assortativity_coefficients.tolist())
```

```python
print("TOPICS RANKING")
rows=[]
for idx, t in enumerate(rank_topics):
    rows.append((idx+1,t[0],assortativity_coefficients[t[0]]))
    
df =pd.DataFrame(rows, columns=["Position","Topic","Assortativity coefficient"])
df
```

```python
print("YOUTUBERS RANKING")
rows=[]
for idx, t in enumerate(rank_youtubers):
    rows.append((idx+1,t[0],assortativity_coefficients[t[0]]))
    
df =pd.DataFrame(rows, columns=["Position","Youtuber","Assortativity coefficient"])
df
```

```python
print("BRANDS RANKING")
rows=[]
for idx, t in enumerate(rank_brands):
    rows.append((idx+1,t[0],assortativity_coefficients[t[0]]))
    
df =pd.DataFrame(rows, columns=["Position","Brand","Assortativity coefficient"])
df
```
