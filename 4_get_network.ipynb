```python
first_graph = nx.DiGraph()
```

```python
youtubers_name_set=set()
for youtuber in df_youtubers.values:
    youtubers_name_set.add(youtuber[0])
```

```python
user_name_set= set()
nan_number=0
for comment in df_comments.values:
    if comment[2]!= None:
        user_name_set.add(comment[2])
    else:
        nan_number+=1
```

```python
tot_nodes= list(youtubers_name_set.union(user_name_set))
```

```python
first_graph.add_nodes_from(tot_nodes)
```

```python
youtuber_video_join=df_youtubers.set_index('channel_id').join(df_youtuber_videos.set_index('channel_id'), on='channel_id',lsuffix='_yt', rsuffix='_ytvd')
```

```python
youtuber_video_join = youtuber_video_join[youtuber_video_join.columns[[0,2]]]
```

```python
youtuber_video_join = youtuber_video_join.set_index('video_id').join(df_videos.set_index('video_id'),on='video_id',lsuffix='_yt', rsuffix='_ytvd')
```

```python
youtuber_video_join
```

```python
youtuber_comments = youtuber_video_join.join(df_comments.set_index('video_id'),on='video_id',lsuffix='_ytvd', rsuffix='_comm')
```

```python
youtuber_comments
```

```python
#aggiungi counter
edges = []
for idx_for,row in enumerate(youtuber_comments.values):
    print('Comment: {}/{}'.format(idx_for,len(youtuber_comments.values)))
    clear_output(wait=True)
    try:
        youtuber = row[0]
        user = row[5]
        arr = [(item[0][0],item[0][1]) for item in edges]
        if (user,youtuber) not in arr:
            edges.append([(user,youtuber),{'video_category':set(row[1]),'topics':set(row[2]),'brands':set(row[3]),'sentiments':[row[6]],'mean_sentiment':np.mean([row[6]])}])
        else:
            idx = arr.index((user,youtuber))
            edges[idx][1]['video_category'].add(row[1])
            edges[idx][1]['topics'].add(row[2])
            edges[idx][1]['brands'].add(row[3])
            edges[idx][1]['sentiments'].append(row[6])
            edges[idx][1]['mean_sentiment'] = np.mean(edges[idx][1]['sentiments'])
    except:
        print('NaN values')
```
