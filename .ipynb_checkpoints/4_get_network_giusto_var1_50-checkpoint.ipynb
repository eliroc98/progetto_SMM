```python
import pandas as pd

import os

import numpy as np

import networkx as nx

import matplotlib.pyplot as plt

from IPython.display import clear_output
```

```python
df_comments = pd.read_csv('data/get_comment_threads_output/comments_thread_sentiment_reduced.csv')
df_videos = pd.read_csv('data/get_videos_details/videos_details_computed.csv')
df_youtuber_videos = pd.read_csv('data/get_youtuber_videos_output/youtuber_videos.csv')
df_youtubers = pd.read_csv('data/get_youtubers_output/youtubers.csv')
```

```python
user_name_set= set()
nan_number=0
for comment in df_comments.values:
    if comment[2]!= None:
        user_name_set.add(comment[2])
    else:
        nan_number+=1
```

```python
youtuber_video_join=df_youtubers.set_index('channel_id').join(df_youtuber_videos.set_index('channel_id'), on='channel_id',lsuffix='_yt', rsuffix='_ytvd')
```

```python
youtuber_video_join = youtuber_video_join[youtuber_video_join.columns[[0,2]]]
```

```python
youtuber_video_join = youtuber_video_join.set_index('video_id').join(df_videos.set_index('video_id'),on='video_id',lsuffix='_yt', rsuffix='_ytvd')
```

```python
youtuber_video_join
```

```python
youtuber_comments = youtuber_video_join.join(df_comments.set_index('video_id'),on='video_id',lsuffix='_ytvd', rsuffix='_comm')
```

```python
youtuber_comments  = (youtuber_comments[youtuber_comments.columns[[0,2,3,4,6,7]]]).dropna()
youtuber_comments
```

```python
edges = []
```

```python
#{'video_category':[row[1]],'topics':[row[2]],'brands':[row[3]],'youtubers':row[4],'sentiments':[row[6]],'mean_sentiment':np.mean([row[6]])}
#i duplicati li guardo quando metto gli edge del grafo (se esiste giÃ  l'edge, aggiungo altri dettagli) + media del peso
#il mean sentiment lo valuterÃ² dopo
i = 1
for idx_for,row_1 in enumerate(youtuber_comments.values):
    video_idxs_not_link = []
    video_idxs_link = dict()
    matches_topics = set([item.strip()[1:-1] for item in row_1[1][1:-1].split(',')])
    matches_brands = set([item.strip()[1:-1] for item in row_1[2][1:-1].split(',')])
    matches_youtubers = set([item.strip()[1:-1] for item in row_1[3][1:-1].split(',')])
    matches_youtubers.add(row_1[0])
    print('___________________________________________________________')
    print('Ottenuti dati per commento {}'.format(idx_for))
    for idx_row_2 , row_2 in enumerate(youtuber_comments.values[i:]):
        if youtuber_comments.index[idx_row_2 + i] not in video_idxs_not_link:
            if youtuber_comments.index[idx_row_2 + i] in video_idxs_link:
                w = video_idxs_link[youtuber_comments.index[idx_row_2 + i]][0]
                details = video_idxs_link[youtuber_comments.index[idx_row_2 + i]][1]
                if row_1[5] == row_2[5]:
                    if row_1[5] == 1:
                        details['sentiment'] = 1
                    else:
                        details['sentiment'] = 0
                else:
                    details['sentiment'] = -1
                edges.append([(row_1[4],row_2[4],w),details])
            else:
                print('-------------------------------------------------------')
                print('Comment: {}-{}/{}'.format(idx_for,idx_row_2 + i,len(youtuber_comments.values[i:])))
                link = False
                details = {'youtubers':[],'topics':[],'brands':[], 'sentiment':0}
                topics_ = set([item.strip()[1:-1] for item in row_2[1][1:-1].split(',')])
                brands_ = set([item.strip()[1:-1] for item in row_2[2][1:-1].split(',')])
                youtubers_ = set([item.strip()[1:-1] for item in row_2[3][1:-1].split(',')])
                youtubers_.add(row_2[0])
                print('Ottenuti dati per commento {}'.format(idx_row_2 + i))
                common_topics = matches_topics.intersection(topics_)
                common_brands = matches_brands.intersection(brands_)
                common_youtubers = matches_youtubers.intersection(youtubers_)
                if len(common_topics)>0:
                    link = True
                    [details['topics'].append(item) for item in common_topics]
                if len(common_brands)>0:
                    link = True
                    [details['brands'].append(item) for item in common_brands]
                if len(common_youtubers)>0:
                    link = True
                    [details['youtubers'].append(item) for item in common_youtubers]

            #CODICI SENTIMENT: 1-->ENTRAMBI 1; 0-->ENTRAMBI 0; -1--> OPINIONI DIVERSE
                if row_1[5] == row_2[5]:
                    if row_1[5] == 1:
                        details['sentiment'] = 1
                    else:
                        details['sentiment'] = 0
                else:
                    details['sentiment'] = -1

                if link:
                    print('link')
                    union_topics = matches_topics.union(topics_)
                    union_brands = matches_brands.union(brands_)
                    union_youtubers = matches_youtubers.union(youtubers_)
                    w = (len(common_topics)+len(common_brands)+len(common_youtubers))/(len(union_topics)+len(union_brands)+len(union_youtubers))
                    video_idxs_link.update({youtuber_comments.index[idx_row_2 + i]:[w, details]})
                    edges.append([(row_1[4],row_2[4],w),details])
                else:
                    print('not link')
                    video_not_link = youtuber_comments.index[idx_row_2 + i]
                    print(video_not_link)
                    video_idxs_not_link.append(video_not_link)
                clear_output(wait=True)
    i += 1
```

```python
np.save('make_network_output/edges_var1_100.npy', edges)
```
